```python
def get_bumped_and_inferred(dim,p_low,p_high,step,num_of_runs):
    global maze_size
    probabilities = np.arange(p_low, p_high, step)
    probabilities = np.round(probabilities,3)
    bumped_dict ={}
    inferred_dict = {}
    
    print("probabilities: ",probabilities)
    for prob in probabilities:
        for i in range(num_of_runs):
            start = time.time()
#             print("started for ", prob, " run: ",i)
            initialize_all_matrices(prob)
            _,_,_,path = astar((0,0,0),block_unblock_matrix)
#             print("path a*:",path)
            if len(path) == 0:
                continue
            else:  
                shortest_path_on_discovered_gridworld,traj_length,num_cells_processed,plan_time,bumped,inferred = agent3()
                
                if prob in bumped_dict.keys():
                    bumped_dict[prob].append(bumped)
                else:
                    bumped_dict[prob] = list()
                    bumped_dict[prob].append(bumped)
                
                if prob in inferred_dict.keys():
                    inferred_dict[prob].append(inferred)
                else:
                    inferred_dict[prob] = list()
                    inferred_dict[prob].append(inferred)

            
        print("Finished for probability: ",prob," run: ",i,"\nbumped value: ",sum(bumped_dict[prob])/len(bumped_dict[prob]))
        print("\ninferred value: ",round(sum(inferred_dict[prob])/len(inferred_dict[prob]),10))
    
    #Density vs Bumped:
    filename = "./results/bumped_"+str(np.random.randint(1000000))+".txt"
    f = open(filename, "w")
    f.write(str(dim)+str(",")+str(num_of_runs)+"\n")
    for key in bumped_dict.keys():
        value = bumped_dict[key]
        f.write(str(key)+ str(",")+str(sum(value)/len(value))+"\n")
    print("done. file: ",filename)
    
    #Density vs Inferred:
    filename = "./results/inferred_"+str(np.random.randint(1000000))+".txt"
    f = open(filename, "w")
    f.write(str(dim)+str(",")+str(num_of_runs)+"\n")
    for key in inferred_dict.keys():
        value = inferred_dict[key]
        f.write(str(key)+ str(",")+str(sum(value)/len(value))+"\n")
    print("done. file: ",filename)
```

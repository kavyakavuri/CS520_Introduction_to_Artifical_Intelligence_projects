```python
def calculateAllValues(dim,p_low,p_high,step,num_of_runs):
    global maze_size
    probabilities = np.arange(p_low, p_high, step)
    probabilities = np.round(probabilities,3)
    trajectory_length ={}
    num_cells = {}
    avg_length = {}
    avg_length_2 = {}
    planning_time = {}
    
    print("probabilities: ",probabilities)
    for prob in probabilities:
        for i in range(num_of_runs):
            start = time.time()
            print("started for ", prob, " run: ",i)
            initialize(dim,prob)
            _,_,_,path = astar((0,0,0),block_unblock_matrix)
#             print("path a*:",path)
            if len(path) == 0:
                continue
            else:
                    
                traj_length,num_cells_processed,shortest_path_on_discovered_gridworld,total_planning_time = repeated_astarq6()
                
                if prob in trajectory_length.keys():
                    trajectory_length[prob].append(traj_length)
                else:
                    trajectory_length[prob] = list()
                    trajectory_length[prob].append(traj_length)
                
                if prob in avg_length.keys():
                    avg_length[prob].append(traj_length/len(shortest_path_on_discovered_gridworld))
                else:
                    avg_length[prob] = list()
                    avg_length[prob].append(traj_length/len(shortest_path_on_discovered_gridworld))
                    
                if prob in avg_length_2.keys():
                    avg_length_2[prob].append(len(shortest_path_on_discovered_gridworld)/len(path))
                else:
                    avg_length_2[prob] = list()
                    avg_length_2[prob].append(len(shortest_path_on_discovered_gridworld)/len(path))
                    
                if prob in num_cells.keys():
                    num_cells[prob].append(num_cells_processed)
                else:
                    num_cells[prob] = list()
                    num_cells[prob].append(num_cells_processed)
                    
                if prob in planning_time.keys():
                    planning_time[prob].append(total_planning_time)
                else:
                    planning_time[prob] = list()
                    planning_time[prob].append(total_planning_time)
             
            time_taken = time.time() - start
            print("Finished for probability: ",prob," run: ",i, " time taken: ",time_taken)
            
    #Density vs Avg Trajectory Length:
    filename = "./results/allValues/agent1/avg_traj_"+str(np.random.randint(1000000))+".txt"
    f = open(filename, "w")
    f.write("agent1"+str(dim)+str(",")+str(num_of_runs)+"\n")
    for key in trajectory_length.keys():
        value = trajectory_length[key]
        f.write(str(key)+ str(",")+str(sum(value)/len(value))+"\n")
    print("done. file: ",filename)
    
    #Density VS Average (Length of Trajectory/length of shortest path on final discovered gridworld)
    filename = "./results/allValues/agent1/avg_len_1_"+str(np.random.randint(1000000))+".txt"
    f = open(filename, "w")
    f.write("agent1"+str(dim)+str(",")+str(num_of_runs)+"\n")
    for key in avg_length.keys():
        value = avg_length[key]
        f.write(str(key)+ str(",")+str(sum(value)/len(value))+"\n")
    print("done. file: ",filename)
    
    #Density VS Average (Length of shortest path on final discovered gridworld / Length of shortest path in Full Gridworld)
    filename = "./results/allValues/agent1/avg_len_2_"+str(np.random.randint(1000000))+".txt"
    f = open(filename, "w")
    f.write("agent1"+str(dim)+str(",")+str(num_of_runs)+"\n")
    for key in avg_length_2.keys():
        value = avg_length_2[key]
        f.write(str(key)+ str(",")+str(sum(value)/len(value))+"\n")
    print("done. file: ",filename)
    
    #Density VS Average Number of Cells Processed
    filename = "./results/allValues/agent1/avg_cells_"+str(np.random.randint(1000000))+".txt"
    f = open(filename, "w")
    f.write("agent1"+str(dim)+str(",")+str(num_of_runs)+"\n")
    for key in num_cells.keys():
        value = num_cells[key]
        f.write(str(key)+ str(",")+str(sum(value)/len(value))+"\n")
    print("done. file: ",filename)
    
    #Density Vs Planning Time
    filename = "./results/allValues/agent1/planning_time_"+str(np.random.randint(1000000))+".txt"
    f = open(filename, "w")
    f.write("agent1"+str(dim)+str(",")+str(num_of_runs)+"\n")
    for key in planning_time.keys():
        value = planning_time[key]
        f.write(str(key)+ str(",")+str(sum(value)/len(value))+"\n")
    print("done. file: ",filename)
```

```python
def get_bumped_q6(dim,p_low,p_high,step,num_of_runs):
    
    probabilities = np.arange(p_low, p_high, step)
    probabilities = np.round(probabilities,3)
    bumped_dict ={}
    
    print("probabilities: ",probabilities)
    for prob in probabilities:
        for i in range(num_of_runs):
#             print("started for ", prob, " run: ",i)
            initialize(dim,prob)
            _,_,_,path = astar((0,0,0),block_unblock_matrix)
#             print("path a*:",path)
            if len(path) == 0:
                continue
            else:
                    
                traj_length,num_cells_processed,shortest_path_on_discovered_gridworld,total_planning_time,bumped = repeated_astarq6()
                
                if prob in bumped_dict.keys():
                    bumped_dict[prob].append(bumped)
                else:
                    bumped_dict[prob] = list()
                    bumped_dict[prob].append(bumped)
        print("Finished for probability: ",prob," run: ",i,"\nbumped value: ",sum(bumped_dict[prob])/len(bumped_dict[prob]))
    
    #Density vs Bumped:
    filename = "./results/allValues/agent1/bumped_"+str(np.random.randint(1000000))+".txt"
    f = open(filename, "w")
    f.write(str(dim)+str(",")+str(num_of_runs)+"\n")
    for key in bumped_dict.keys():
        value = bumped_dict[key]
        f.write(str(key)+ str(",")+str(sum(value)/len(value))+"\n")
    print("done. file: ",filename)
```
